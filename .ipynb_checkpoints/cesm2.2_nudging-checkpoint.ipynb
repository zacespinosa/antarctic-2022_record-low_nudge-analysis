```python
def spatial_era5_plots(df, title, sp):
    fig, ax = plot_stationary_sp()

    # Plot siconc filled contours
    img = ax.pcolormesh(
        df.lon,
        df.lat,
        df,
        vmin=-10,
        vmax=10,
        transform=ccrs.PlateCarree(), 
        cmap="RdBu",
    )
    cbar2 = fig.colorbar(img, ax=ax, orientation="horizontal")
    ax.set_title(title)
    fig.set_size_inches(8, 8)
```
